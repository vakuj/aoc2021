```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
fish_fun = lambda x,c,t:x*t+c
fishes = [300, 600, 1200, 2527, 5146, 10353, 20957, 42065, 84954, 170249, 343441, 687689,
              1386348, 2774969, 5591732, 11191737, 22544358, 45125437, 90872848, 181922671, 366253298]
fishes = np.log(fishes)
days = [0, 6, 15, 24, 32, 40, 48, 56, 64, 72, 80,
        88, 96, 104, 112, 120, 128, 136, 144, 152, 160]

popt,_ = curve_fit(fish_fun,days,fishes)
print(popt)
day_ext = np.linspace(0,160,num=160)
print(np.exp(fish_fun(256,popt[0],popt[1])))
fig,ax =plt.subplots(1,1,figsize=(10,10))
ax.plot(days, fishes,label='input')
ax.plot(day_ext, fish_fun(day_ext,popt[0],popt[1]),'r-.',label='fitted')
ax.grid()
ax.legend()
plt.show()
```
